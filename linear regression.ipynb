```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
l = 1000
x = np.arange(l) / l
y = 10 * x + np.random.standard_normal(l)
plt.scatter(x, y, s=2)
```

```python
def grad(x, y, theta):
  return np.array([((theta @ x - y)/(2 * len(y))).sum(), (((theta @ x - y) @ (x[1])) /len(y)).sum()])
```

```python
theta = np.array([0, 0])
p = np.ones(1000)
p.shape = (1000, 1)
x.shape = (1000, 1)
x = np.concatenate((p, x), axis=1).transpose()
```

```python
def sgd(x, y, n = 1000, alfa = 0.1):
  theta = np.array([0, 0])
  p = np.ones(1000)
  p.shape = (1000, 1)
  x.shape = (1000, 1)
  x = np.concatenate((p, x), axis=1).transpose()
  k = 0
  while k < n:
    theta = theta - alfa * grad(x, y, theta)
    k+=1
  return theta
```

```python
param = sgd(x, y)
param
```

```python
kl = x * param[1] + param[0]
```

```python
plt.scatter(x, kl, s=2)
plt.scatter(x, y, s=2)
```
