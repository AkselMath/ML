```python
def signs(image : np.array, where : str):
    if where == 'indoor':
        asd = 1
    else:
        asd = 0
    shape_img = image.shape[0] * image.shape[1]
    red_intensity = image[:,:,0].sum() / shape_img
    green_intensity = image[:,:,1].sum() / shape_img
    blue_intensity = image[:,:,2].sum() / shape_img
        
    red_deviation = np.power(image[:,:,0] - red_intensity, 2).sum() / shape_img
    green_deviation = np.power(image[:,:,1] - green_intensity, 2).sum() / shape_img
    blue_deviation = np.power(image[:,:,2] - blue_intensity, 2).sum() / shape_img
                              
    return [red_intensity, green_intensity, blue_intensity, red_deviation, green_deviation, blue_deviation, asd]
```

```python
def signs_half(image : np.array, where : str):
    if where == 'indoor':
        asd = 1
    else:
        asd = 0
    
    upper_half = image[:int(image.shape[0]/2)]
    upper_half_shape_img = upper_half.shape[0] * upper_half.shape[1]
    
    upper_half_red_intensity = upper_half[:,:,0].sum() / upper_half_shape_img
    upper_half_green_intensity = upper_half[:,:,1].sum() / upper_half_shape_img
    upper_half_blue_intensity = upper_half[:,:,2].sum() / upper_half_shape_img
        
    upper_half_red_deviation = np.power(upper_half[:,:,0] - upper_half_red_intensity, 2).sum() / upper_half_shape_img
    upper_half_green_deviation = np.power(upper_half[:,:,1] - upper_half_green_intensity, 2).sum() / upper_half_shape_img
    upper_half_blue_deviation = np.power(upper_half[:,:,2] - upper_half_blue_intensity, 2).sum() / upper_half_shape_img
                              
    
    bottom_half = image[int(image.shape[0]/2):]
    bottom_half_shape_img = bottom_half.shape[0] * bottom_half.shape[1]
    
    bottom_half_red_intensity = bottom_half[:,:,0].sum() / bottom_half_shape_img
    bottom_half_green_intensity = bottom_half[:,:,1].sum() / bottom_half_shape_img
    bottom_half_blue_intensity = bottom_half[:,:,2].sum() / bottom_half_shape_img
        
    bottom_half_red_deviation = np.power(bottom_half[:,:,0] - bottom_half_red_intensity, 2).sum() / bottom_half_shape_img
    bottom_half_green_deviation = np.power(bottom_half[:,:,1] - bottom_half_green_intensity, 2).sum() / bottom_half_shape_img
    bottom_half_blue_deviation = np.power(bottom_half[:,:,2] - bottom_half_blue_intensity, 2).sum() / bottom_half_shape_img
                              
    
    return [upper_half_red_intensity, upper_half_green_intensity, upper_half_blue_intensity, upper_half_red_deviation, upper_half_green_deviation, upper_half_blue_deviation,
            bottom_half_red_intensity, bottom_half_green_intensity, bottom_half_blue_intensity, bottom_half_red_deviation, bottom_half_blue_deviation, asd]
```
