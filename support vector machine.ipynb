```python
import numpy as np
import scipy
import matplotlib.pyplot as plt

from scipy.optimize import minimize
from scipy.optimize import Bounds
from scipy.optimize import LinearConstraint
```

```python
data = np.genfromtxt('svm_data.csv', delimiter=',')
```

```python
y = np.array(data)[:,-1]
x = np.array(data)[:,:-1]
```

```python
def L(lambd : np.array, y : np.array = y, x : np.array = x):
  sum = 0
  for i in range(len(lambd)):
    for j in range(len(lambd)):
      sum += lambd[i] * lambd[j] * y[i] * y[j] * (x[i] @ x[j])
  sum = -sum / 2 + lambd.sum()
  return -sum
```

```python
bounds = Bounds(np.zeros(11), [np.inf, np.inf, np.inf, np.inf, np.inf, np.inf, np.inf, np.inf, np.inf, np.inf, np.inf])

linear_constraint = LinearConstraint ([y], [0], [0])
linear_constraint
```

```python
a = minimize(lambda x : L(x), x0=np.ones(11), constraints=linear_constraint, bounds=bounds)
a
```

```python
lambd = a.x
theta = [0, 0]

for i in range(10):
  theta += x[i] * (lambd * y)[i]
```

```python
zero_i = np.nonzero(lambd)[0]

t = []
for i in zero_i:
  t.append(theta @ x[i])

theta_0 = np.median(y[zero_i] - np.array(t))
```
