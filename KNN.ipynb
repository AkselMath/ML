```python
import numpy as np
import pandas as pd
import cv2
import zipfile
import os

from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import KFold
```

```python
base_dir = '/kaggle/input/cian-datafest-2019/train.zip'
with zipfile.ZipFile(base_dir, 'r') as z:
    z.extractall()
```

```python
img = cv2.imread('./train/indoor/3.jpg') / 255.
img = cv2.resize(img, (10, 10), interpolation = cv2.INTER_AREA)
img = img.reshape(-1)
```

```python
a = []
for i in os.listdir('train/indoor'):
    img = cv2.imread('./train/indoor/' + i) / 255.
    img = cv2.resize(img, (10, 10), interpolation = cv2.INTER_AREA)
    img = img.reshape(-1)
    img = list(img) + [0]
    a.append(img)
    
for i in os.listdir('train/outdoor'):
    img = cv2.imread('./train/outdoor/' + i) / 255.
    img = cv2.resize(img, (10, 10), interpolation = cv2.INTER_AREA)
    img = img.reshape(-1)
    img = list(img) + [1]
    a.append(img)

a = np.array(a)
```

```python
kf = KFold(n_splits=5, shuffle=True, random_state=42)
parameters = {'n_neighbors' : [3, 5, 7, 9]}
neigh = KNeighborsClassifier()

clf = GridSearchCV(neigh, parameters, cv=kf)
```

```python
clf.fit(a[:,:-1], a[:,-1])
```

```python
print(clf.best_params_)
print(clf.best_score_)
```
